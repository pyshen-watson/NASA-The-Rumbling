## This notebook will turn analog data we extract from video into music sheet

```python
import numpy as np

VIDEO_TYPE = 'High'

rhythm = np.load(f'data/rhythm_{VIDEO_TYPE}.npy')
spectrum = np.load(f'data/spectrum_{VIDEO_TYPE}.npy')
```

We use this function to turn analog signal into digital

```python
def list_to_str(l):
    score = ['0' if k==1 else '-' for k in l]
    score = ''.join(score)
    return score
```

## The rhythm

```python
r = (np.log(rhythm + 1e-8)/2).astype(np.int32)
size = r.max()
scores = []

for i in range(size):
    index = (r >= i+1).astype(np.uint8).tolist()
    score = list_to_str(index)
    scores.append(score)
```

```python
print(scores) # This output is used in the music sheet
```

## The spectrum

```python
s = spectrum
median = np.median(s, axis=1)
threshold = np.median(median)
s0 = (s>threshold).sum(axis=1) / s.shape[1]

one = s0 > 0.65
two = np.bitwise_and((s0 > 0.5), (s0 < 0.65))
three = s0 < 0.5
```

```python
print([list_to_str(l) for l in [one, two, three]]) # This output is used in the music sheet
```
